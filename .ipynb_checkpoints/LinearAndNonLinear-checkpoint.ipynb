# Band structure and DOS of three-band system

```python
infinity = 100
T = 300 #Temperature in K
kB = 0.00008617 #Bolztman constant in eV/K
kBT = kB*T

# linear valence parameter
a = 1.3 #valence 2.555
deltav = 0.1 #2*kBT
Dv = deltav/kBT

# linear conduction parameter

b = 1.3 #conduction 2.472
deltac = 0.1 #2*kBT #Conduction -0.08040215
Dc = deltac/kBT

# parabolic parameter
c2 = 2.563 #parabolic 2.563 1/(2*pi**2)
deltah2 = 0.1
Dh2 = deltah2/kBT
E02 = 5 #parabolic bandwidth
Eh2 = E02/kBT

# pudding mold 4th parameter
c4 = 2.563 #puddingmold4th
deltah4 = 0.1
Dh4 = deltah4/kBT
E04 = 5 #parabolic bandwidth
Eh4 = E04/kBT

# pudding mold 6th parameter
c6 = 2.563 #puddingmold6th
deltah6 = 0.1
Dh6 = deltah6/kBT
E06 = 5 #parabolic bandwidth
Eh6 = E06/kBT

def DOS(E,d,n,m):
    ds = zeros(E.size)
    for i in range(E.size):
        if(E[i]>0.1):
            ds[i] = ((1/n)*(1/pi*d**0.5)**d*(E[i]-0.1)**(-1+d/n)) + ((1/m)*(1/pi*d**0.5)**d*(E[i]-0.1)**(-1+d/m))
        elif(-0.1<E[i]<0.1):
            ds[i]=0
        else:
            ds[i] = (1/n)*(1/pi*d**0.5)**d*(-E[i]-0.1)**(-1+d/n)
    return ds

eta = linspace(-6,6,1000)


# Define the size of the entire plot
fig = plt.figure(figsize=(10,8))

# Add two axes objects to create a paneled figure
axBand1 = fig.add_axes([0, 0.8, 0.33, 0.33])
axDOS1 = fig.add_axes([0.42, 0.8, 0.33, 0.33])


axBand2 = fig.add_axes([0, 0.4, 0.33, 0.33])
axDOS2 = fig.add_axes([0.42, 0.4, 0.33, 0.33])


axBand3 = fig.add_axes([0, 0, 0.33, 0.33])
axDOS3 = fig.add_axes([0.42, 0, 0.33, 0.33])


# Edit the major and minor ticks of the x and y axes
axBand1.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axBand1.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')
axDOS1.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axDOS1.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')

axBand1.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axBand1.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axDOS1.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axDOS1.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')

axBand2.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axBand2.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')
axDOS2.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axDOS2.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')

axBand2.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axBand2.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axDOS2.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axDOS2.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')

axBand3.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axBand3.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')
axDOS3.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axDOS3.xaxis.set_tick_params(which='minor', size=6, direction='in', top='on')

axBand3.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axBand3.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axDOS3.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axDOS3.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')


#----------------------------------------------------------------------
# Plotting
#----------------------------------------------------------------------

#----------------------------------------------------------------------
# Parabolic
#----------------------------------------------------------------------

axBand1.plot(eta, c2*eta**2+deltah2, 'r-')
axBand1.plot(eta, b*abs(eta)+deltah2, 'g-.')
axBand1.plot(eta, -a*abs(eta)-deltav, 'b--')
!axBand1.legend(loc = 'upper right')

axDOS1.plot(DOS(eta,3,1,2)/pi,eta, linestyle= 'solid', linewidth=2.0, color='black', label = '$\sigma_v$')
!axDOS1.plot(DOS(eta,3,4)/pi,eta, 'g--', label = '$\sigma_h$')
!axDOS1.plot(DOS(eta,3,6)/pi,eta, 'b--', label = '$\sigma_c$')
!axDOS1.plot(eta, sigma_e2, color = 'darkorange',linewidth=2, label = '$\sigma$')
!axDOS1.legend(loc = 'lower left')

#----------------------------------------------------------------------
# Pudding Mold 4th
#----------------------------------------------------------------------

axBand2.plot(eta, c4*eta**4+deltah2, 'r-')
axBand2.plot(eta, b*abs(eta)+deltah2, 'g-.')
axBand2.plot(eta, -a*abs(eta)-deltav, 'b--')
!axBand2.legend(loc = 'upper right')

axDOS2.plot(DOS(eta,3,1,4)/pi,eta, linestyle= 'solid', linewidth=2.0, color='black', label = '$\sigma_v$')
!axDOS2.plot(eta, sigma_h4, 'g--', label = '$\sigma_h$')
!axDOS2.plot(eta, sigma_c, 'b--', label = '$\sigma_c$')
!axDOS2.plot(eta, sigma_e4, color = 'darkorange',linewidth=2, label = '$\sigma$')
!axDOS2.legend(loc = 'lower left')

#----------------------------------------------------------------------
# Pudding Mold 6th
#----------------------------------------------------------------------

axBand3.plot(eta, c6*eta**6+deltah2, 'r-')
axBand3.plot(eta, b*abs(eta)+deltah2, 'g-.')
axBand3.plot(eta, -a*abs(eta)-deltav, 'b--')
!axBand3.legend(loc = 'upper right')

axDOS3.plot(DOS(eta,3,1,6)/pi,eta, linestyle= 'solid', linewidth=2.0, color='black', label = '$\sigma_v$')
!axDOS3.plot(eta, sigma_h6, 'g--', label = '$\sigma_h$')
!axDOS3.plot(eta, sigma_c, 'b--', label = '$\sigma_c$')
!axDOS3.plot(eta, sigma_e6, color = 'darkorange',linewidth=2, label = '$\sigma$')
!axDOS3.legend(loc = 'lower left')


#######################################################################
# Format Axis Labels and Ticks
#######################################################################

axBand1.axhline(0, color= 'grey', linestyle='-', linewidth=1)
axBand1.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axBand1.xaxis.set_major_locator(mpl.ticker.MultipleLocator(4.0))
axBand1.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1.0))
axBand1.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axBand1.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.2))
axBand1.set_xlim(-4,4); axBand1.set_ylim(-2,2)
axBand1.set_ylabel(r'$E$ ($eV$)', labelpad=2, fontsize = 24)

axDOS1.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axDOS1.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axDOS1.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
axDOS1.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axDOS1.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axDOS1.set_xlim(0,0.3); axDOS1.set_ylim(-2,2)

axBand2.axhline(0, color= 'grey', linestyle='-', linewidth=1)
axBand2.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axBand2.xaxis.set_major_locator(mpl.ticker.MultipleLocator(4.0))
axBand2.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1.0))
axBand2.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axBand2.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.2))
axBand2.set_xlim(-4,4); axBand2.set_ylim(-2,2)
axBand2.set_ylabel(r'$E$ ($eV$)', labelpad=0, fontsize = 24)

axDOS2.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axDOS2.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axDOS2.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
axDOS2.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axDOS2.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axDOS2.set_xlim(0,0.3); axDOS2.set_ylim(-2,2)

axBand3.axhline(0, color= 'grey', linestyle='-', linewidth=1)
axBand3.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axBand3.xaxis.set_major_locator(mpl.ticker.MultipleLocator(4.0))
axBand3.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(1.0))
axBand3.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axBand3.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.2))
axBand3.set_xlim(-4,4); axBand3.set_ylim(-2,2)
axBand3.set_ylabel(r'$E$ ($eV$)', labelpad=0, fontsize = 24)
axBand3.set_xlabel('$k$', fontsize = 24)

axDOS3.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axDOS3.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.1))
axDOS3.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.05))
axDOS3.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axDOS3.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axDOS3.set_xlim(0,0.3); axDOS3.set_ylim(-2,2)
axDOS3.set_xlabel('DOS',fontsize = 24)


## Give panel labels
axBand1.text(  -5.8, 1.9, '(a)', fontsize=24)
axDOS1.text(-0.068, 1.9, '(b)', fontsize=24)

axBand2.text(  -5.8, 1.9, '(c)', fontsize=24)
axDOS2.text(-0.068, 1.9, '(d)', fontsize=24)

axBand3.text(  -5.8, 1.9, '(e)', fontsize=24)
axDOS3.text(-0.068, 1.9, '(f)', fontsize=24)

plt.savefig('Fig1.pdf', bbox_inches='tight')
```

```python
infinity = 100
T = 300 #Temperature in K
kB = 0.00008617 #Bolztman constant in eV/K
kBT = kB*T

## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return ( -(x**3)/(1+exp(x)) - 3*(x**2)*log(1+exp(-x)) + 6*x*plog(2,-exp(-x)) + 6*plog(3,-exp(-x)) )

def L0v(xx,a,Dv) :
    return a**2*F0v(xx+Dv)
def L1v(xx,a,Dv):
    return a**2*F1v(xx+Dv)
def L2v(xx,a,Dv):
    return a**2*F2v(xx+Dv)

## conduction band
def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3c(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x))
            - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )


def L0c(xx,b,Dc) :
    return b**2*F0c(xx-Dc)
def L1c(xx,b,Dc):
    return b**2*F1c(xx-Dc)
def L2c(xx,b,Dc):
    return b**2*F2c(xx-Dc)

## parabolic band
def F0h2(x):
    return exp(x)/(1+exp(x))
def F1h2(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2h2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3h2(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x))
            - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

def L0h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F1h2(xx-Dh2)+(xx-Dh2)*F0h2(xx-Dh2)-F1h2(xx-Eh2)-(xx-Dh2)*F0h2(xx-Eh2))
def L1h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F2h2(xx-Dh2)+(xx-Dh2)*F1h2(xx-Dh2)-F2h2(xx-Eh2)-(xx-Dh2)*F1h2(xx-Eh2))
def L2h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F3h2(xx-Dh2)+(xx-Dh2)*F2h2(xx-Dh2)-F3h2(xx-Eh2)-(xx-Dh2)*F2h2(xx-Eh2))

## Pudding mold 4th
def L0h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I

def L1h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*(x-xx)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I

def L2h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*(x-xx)**2*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I



# Pudding mold 6th
def L0h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I

def L1h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*(x-xx)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I

def L2h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*(x-xx)**2*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I


#TE Properties

def Sc(x,b,Dc):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1c(x[i],b,Dc) / L0c(x[i],b,Dc)
    return seeb
def sigma_c(x,b,Dc):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0c(x[i],b,Dc)
    return sgm
def kappa_c(x,b,Dc):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2c(x[i],b,Dc)-(L1c(x[i],b,Dc)**2/L0c(x[i],b,Dc)))
    return kpp

def Sv(x,a,Dv):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1v(x[i],a,Dv) / L0v(x[i],a,Dv)
    return seeb
def sigma_v(x,a,Dv):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0v(x[i],a,Dv)
    return sgm
def kappa_v(x,a,Dv):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2v(x[i],a,Dv)-(L1v(x[i],a,Dv)**2/L0v(x[i],a,Dv)))
    return kpp

def Sh2(x,c2,Dh2,Eh2):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h2(x[i],c2,Dh2,Eh2) / L0h2(x[i],c2,Dh2,Eh2)
    return seeb
def sigma_h2(x,c2,Dh2,Eh2):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h2(x[i],c2,Dh2,Eh2)
    return sgm
def kappa_h2(x,c2,Dh2,Eh2):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h2(x[i],c2,Dh2,Eh2)-(L1h2(x[i],c2,Dh2,Eh2)**2/L0h2(x[i],c2,Dh2,Eh2)))
    return kpp

def Sh4(x,c4,Dh4,Eh4):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h4(x[i],c4,Dh4,Eh4) / L0h4(x[i],c4,Dh4,Eh4)
    return seeb
def sigma_h4(x,c4,Dh4,Eh4):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h4(x[i],c4,Dh4,Eh4)
    return sgm
def kappa_h4(x,c4,Dh4,Eh4):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h4(x[i],c4,Dh4,Eh4)-(L1h4(x[i],c4,Dh4,Eh4)**2/L0h4(x[i],c4,Dh4,Eh4)))
    return kpp

def Sh6(x,c6,Dh6,Eh6):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h6(x[i],c6,Dh6,Eh6) / L0h6(x[i],c6,Dh6,Eh6)
    return seeb
def sigma_h6(x,c6,Dh6,Eh6):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h6(x[i],c6,Dh6,Eh6)
    return sgm
def kappa_h6(x,c6,Dh6,Eh6):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h6(x[i],c6,Dh6,Eh6)-(L1h6(x[i],c6,Dh6,Eh6)**2/L0h6(x[i],c6,Dh6,Eh6)))
    return kpp

def sigma2(x,a,b,c2,Dv,Dc,Dh2,Eh2):
    return sigma_v(x,a,Dv)+sigma_c(x,b,Dc)+sigma_h2(x,c2,Dh2,Eh2)
def Seebeck2(x,a,b,c2,Dv,Dc,Dh2,Eh2):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_c(x,b,Dc)*Sc(x,b,Dc)+sigma_h2(x,c2,Dh2,Eh2)*Sh2(x,c2,Dh2,Eh2))/sigma2(x,a,b,c2,Dv,Dc,Dh2,Eh2)
def kappae2(x,a,b,c2,Dv,Dc,Dh2,Eh2):
    return ((sigma_c(x,b,Dc)*sigma_v(x,a,Dv)*(Sc(x,b,Dc)-Sv(x,a,Dv))**2)/(sigma_c(x,b,Dc)+sigma_v(x,a,Dv))
            + (sigma_c(x,b,Dc)*sigma_h2(x,c2,Dh2,Eh2)*(Sc(x,b,Dc)-Sh2(x,c2,Dh2,Eh2))**2)/(sigma_c(x,b,Dc)+sigma_h2(x,c2,Dh2,Eh2))
            + (sigma_h2(x,c2,Dh2,Eh2)*sigma_v(x,a,Dv)*(Sh2(x,c2,Dh2,Eh2)-Sv(x,a,Dv))**2)/(sigma_h2(x,c2,Dh2,Eh2)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_c(x,b,Dc)+kappa_h2(x,c2,Dh2,Eh2)))
def PF2(x,a,b,c2,Dv,Dc,Dh2,Eh2):
    return sigma2(x,a,b,c2,Dv,Dc,Dh2,Eh2)*Seebeck2(x,a,b,c2,Dv,Dc,Dh2,Eh2)**2
def ZT2(x,a,b,c2,Dv,Dc,Dh2,Eh2):
    return (sigma2(x,a,b,c2,Dv,Dc,Dh2,Eh2)*Seebeck2(x,a,b,c2,Dv,Dc,Dh2,Eh2)**2)/kappae2(x,a,b,c2,Dv,Dc,Dh2,Eh2)

def sigma4(x,a,b,c4,Dv,Dc,Dh4,Eh4):
    return sigma_v(x,a,Dv)+sigma_c(x,b,Dc)+sigma_h4(x,c4,Dh4,Eh4)
def Seebeck4(x,a,b,c4,Dv,Dc,Dh4,Eh4):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_c(x,b,Dc)*Sc(x,b,Dc)+sigma_h4(x,c4,Dh4,Eh4)*Sh4(x,c4,Dh4,Eh4))/sigma4(x,a,b,c4,Dv,Dc,Dh4,Eh4)
def kappae4(x,a,b,c4,Dv,Dc,Dh4,Eh4):
    return ((sigma_c(x,b,Dc)*sigma_v(x,a,Dv)*(Sc(x,b,Dc)-Sv(x,a,Dv))**2)/(sigma_c(x,b,Dc)+sigma_v(x,a,Dv))
            + (sigma_c(x,b,Dc)*sigma_h4(x,c4,Dh4,Eh4)*(Sc(x,b,Dc)-Sh4(x,c4,Dh4,Eh4))**2)/(sigma_c(x,b,Dc)+sigma_h4(x,c4,Dh4,Eh4))
            + (sigma_h4(x,c4,Dh4,Eh4)*sigma_v(x,a,Dv)*(Sh4(x,c4,Dh4,Eh4)-Sv(x,a,Dv))**2)/(sigma_h4(x,c4,Dh4,Eh4)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_c(x,b,Dc)+kappa_h4(x,c4,Dh4,Eh4)))
def PF4(x,a,b,c4,Dv,Dc,Dh4,Eh4):
    return sigma4(x,a,b,c4,Dv,Dc,Dh4,Eh4)*Seebeck4(x,a,b,c4,Dv,Dc,Dh4,Eh4)**2
def ZT4(x,a,b,c4,Dv,Dc,Dh4,Eh4):
    return (sigma4(x,a,b,c4,Dv,Dc,Dh4,Eh4)*Seebeck4(x,a,b,c4,Dv,Dc,Dh4,Eh4)**2)/kappae4(x,a,b,c4,Dv,Dc,Dh4,Eh4)

def sigma6(x,a,b,c6,Dv,Dc,Dh6,Eh6):
    return sigma_v(x,a,Dv)+sigma_c(x,b,Dc)+sigma_h6(x,c6,Dh6,Eh6)
def Seebeck6(x,a,b,c6,Dv,Dc,Dh6,Eh6):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_c(x,b,Dc)*Sc(x,b,Dc)+sigma_h6(x,c6,Dh6,Eh6)*Sh6(x,c6,Dh6,Eh6))/sigma6(x,a,b,c6,Dv,Dc,Dh6,Eh6)
def kappae6(x,a,b,c6,Dv,Dc,Dh6,Eh6):
    return ((sigma_c(x,b,Dc)*sigma_v(x,a,Dv)*(Sc(x,b,Dc)-Sv(x,a,Dv))**2)/(sigma_c(x,b,Dc)+sigma_v(x,a,Dv))
            + (sigma_c(x,b,Dc)*sigma_h6(x,c6,Dh6,Eh6)*(Sc(x,b,Dc)-Sh6(x,c6,Dh6,Eh6))**2)/(sigma_c(x,b,Dc)+sigma_h6(x,c6,Dh6,Eh6))
            + (sigma_h6(x,c6,Dh6,Eh6)*sigma_v(x,a,Dv)*(Sh6(x,c6,Dh6,Eh6)-Sv(x,a,Dv))**2)/(sigma_h6(x,c6,Dh6,Eh6)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_c(x,b,Dc)+kappa_h6(x,c6,Dh6,Eh6)))
def PF6(x,a,b,c6,Dv,Dc,Dh6,Eh6):
    return sigma6(x,a,b,c6,Dv,Dc,Dh6,Eh6)*Seebeck6(x,a,b,c6,Dv,Dc,Dh6,Eh6)**2
def ZT6(x,a,b,c6,Dv,Dc,Dh6,Eh6):
    return (sigma6(x,a,b,c6,Dv,Dc,Dh6,Eh6)*Seebeck6(x,a,b,c6,Dv,Dc,Dh6,Eh6)**2)/kappae6(x,a,b,c6,Dv,Dc,Dh6,Eh6)

# Dirac bands
def sigmaD(x,a,b,Dv,Dc):
    return sigma_v(x,a,Dv)+sigma_c(x,b,Dc)
def SeebeckD(x,a,b,Dv,Dc):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_c(x,b,Dc)*Sc(x,b,Dc))/sigmaD(x,a,b,Dv,Dc)
def kappaeD(x,a,b,Dv,Dc):
    return ((sigma_c(x,b,Dc)*sigma_v(x,a,Dv)*(Sc(x,b,Dc)-Sv(x,a,Dv))**2)/(sigma_c(x,b,Dc)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_c(x,b,Dc)))
def PFD(x,a,b,Dv,Dc):
    return sigmaD(x,a,b,Dv,Dc)*SeebeckD(x,a,b,Dv,Dc)**2
def ZTD(x,a,b,Dv,Dc):
    return (sigmaD(x,a,b,Dv,Dc)*SeebeckD(x,a,b,Dv,Dc)**2)/kappaeD(x,a,b,Dv,Dc)

npoint = 101
eta = linspace(-10, 10, npoint) 
```

# TE properties of two-band system

```python
infinity = 100
T = 300 #Temperature in K
kB = 0.00008617 #Bolztman constant in eV/K
kBT = kB*T

## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return ( -(x**3)/(1+exp(x)) - 3*(x**2)*log(1+exp(-x)) + 6*x*plog(2,-exp(-x)) + 6*plog(3,-exp(-x)) )

def L0v(xx,a,Dv) :
    return a**2*F0v(xx+Dv)
def L1v(xx,a,Dv):
    return a**2*F1v(xx+Dv)
def L2v(xx,a,Dv):
    return a**2*F2v(xx+Dv)

## parabolic band
def F0h2(x):
    return exp(x)/(1+exp(x))
def F1h2(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2h2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3h2(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x))
            - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

def L0h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F1h2(xx-Dh2)+(xx-Dh2)*F0h2(xx-Dh2)-F1h2(xx-Eh2)-(xx-Dh2)*F0h2(xx-Eh2))
def L1h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F2h2(xx-Dh2)+(xx-Dh2)*F1h2(xx-Dh2)-F2h2(xx-Eh2)-(xx-Dh2)*F1h2(xx-Eh2))
def L2h2(xx,c2,Dh2,Eh2) :
    return 4*c2*kBT*(F3h2(xx-Dh2)+(xx-Dh2)*F2h2(xx-Dh2)-F3h2(xx-Eh2)-(xx-Dh2)*F2h2(xx-Eh2))

## Pudding mold 4th
def L0h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I

def L1h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*(x-xx)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I

def L2h4(xx,c4,Dh4,Eh4):
    def integrand(x,xx):
        return ((x-Dh4)**(3/2)*(x-xx)**2*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh4,Eh4,args=(xx))
    return 16*c4**(1/2)*(kBT)**(3/2)*I



# Pudding mold 6th
def L0h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I

def L1h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*(x-xx)*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I

def L2h6(xx,c6,Dh6,Eh6):
    def integrand(x,xx):
        return ((x-Dh6)**(5/3)*(x-xx)**2*exp(x-xx)/(exp(x-xx)+1)**2)
    I, error = quad(integrand,Dh6,Eh6,args=(xx))
    return 36*c6**(1/3)*(kBT)**(5/3)*I


#TE Properties
def Sv(x,a,Dv):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1v(x[i],a,Dv) / L0v(x[i],a,Dv)
    return seeb
def sigma_v(x,a,Dv):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0v(x[i],a,Dv)
    return sgm
def kappa_v(x,a,Dv):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2v(x[i],a,Dv)-(L1v(x[i],a,Dv)**2/L0v(x[i],a,Dv)))
    return kpp

def Sh2(x,c2,Dh2,Eh2):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h2(x[i],c2,Dh2,Eh2) / L0h2(x[i],c2,Dh2,Eh2)
    return seeb
def sigma_h2(x,c2,Dh2,Eh2):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h2(x[i],c2,Dh2,Eh2)
    return sgm
def kappa_h2(x,c2,Dh2,Eh2):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h2(x[i],c2,Dh2,Eh2)-(L1h2(x[i],c2,Dh2,Eh2)**2/L0h2(x[i],c2,Dh2,Eh2)))
    return kpp

def Sh4(x,c4,Dh4,Eh4):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h4(x[i],c4,Dh4,Eh4) / L0h4(x[i],c4,Dh4,Eh4)
    return seeb
def sigma_h4(x,c4,Dh4,Eh4):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h4(x[i],c4,Dh4,Eh4)
    return sgm
def kappa_h4(x,c4,Dh4,Eh4):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h4(x[i],c4,Dh4,Eh4)-(L1h4(x[i],c4,Dh4,Eh4)**2/L0h4(x[i],c4,Dh4,Eh4)))
    return kpp

def Sh6(x,c6,Dh6,Eh6):
    seeb = zeros(x.size)
    for i in range(x.size):
        seeb[i] = -L1h6(x[i],c6,Dh6,Eh6) / L0h6(x[i],c6,Dh6,Eh6)
    return seeb
def sigma_h6(x,c6,Dh6,Eh6):
    sgm = zeros(x.size)
    for i in range(x.size):
        sgm[i] = L0h6(x[i],c6,Dh6,Eh6)
    return sgm
def kappa_h6(x,c6,Dh6,Eh6):
    kpp = zeros(x.size)
    for i in range(x.size):
        kpp[i] = (L2h6(x[i],c6,Dh6,Eh6)-(L1h6(x[i],c6,Dh6,Eh6)**2/L0h6(x[i],c6,Dh6,Eh6)))
    return kpp

def sigma2(x,a,c2,Dv,Dh2,Eh2):
    return sigma_v(x,a,Dv)+sigma_h2(x,c2,Dh2,Eh2)
def Seebeck2(x,a,c2,Dv,Dh2,Eh2):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_h2(x,c2,Dh2,Eh2)*Sh2(x,c2,Dh2,Eh2))/sigma2(x,a,c2,Dv,Dh2,Eh2)
def kappae2(x,a,c2,Dv,Dh2,Eh2):
    return ((sigma_h2(x,c2,Dh2,Eh2)*sigma_v(x,a,Dv)*(Sh2(x,c2,Dh2,Eh2)-Sv(x,a,Dv))**2)/(sigma_h2(x,c2,Dh2,Eh2)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_h2(x,c2,Dh2,Eh2)))
def PF2(x,a,c2,Dv,Dh2,Eh2):
    return sigma2(x,a,c2,Dv,Dh2,Eh2)*Seebeck2(x,a,c2,Dv,Dh2,Eh2)**2
def ZT2(x,a,c2,Dv,Dh2,Eh2):
    return (sigma2(x,a,c2,Dv,Dh2,Eh2)*Seebeck2(x,a,c2,Dv,Dh2,Eh2)**2)/kappae2(x,a,c2,Dv,Dh2,Eh2)

def sigma4(x,a,c4,Dv,Dh4,Eh4):
    return sigma_v(x,a,Dv)+sigma_h4(x,c4,Dh4,Eh4)
def Seebeck4(x,a,c4,Dv,Dh4,Eh4):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_h4(x,c4,Dh4,Eh4)*Sh4(x,c4,Dh4,Eh4))/sigma4(x,a,c4,Dv,Dh4,Eh4)
def kappae4(x,a,c4,Dv,Dh4,Eh4):
    return ((sigma_h4(x,c4,Dh4,Eh4)*sigma_v(x,a,Dv)*(Sh4(x,c4,Dh4,Eh4)-Sv(x,a,Dv))**2)/(sigma_h4(x,c4,Dh4,Eh4)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_h4(x,c4,Dh4,Eh4)))
def PF4(x,a,c4,Dv,Dh4,Eh4):
    return sigma4(x,a,c4,Dv,Dh4,Eh4)*Seebeck4(x,a,c4,Dv,Dh4,Eh4)**2
def ZT4(x,a,c4,Dv,Dh4,Eh4):
    return (sigma4(x,a,c4,Dv,Dh4,Eh4)*Seebeck4(x,a,c4,Dv,Dh4,Eh4)**2)/kappae4(x,a,c4,Dv,Dh4,Eh4)

def sigma6(x,a,c6,Dv,Dh6,Eh6):
    return sigma_v(x,a,Dv)+sigma_h6(x,c6,Dh6,Eh6)
def Seebeck6(x,a,c6,Dv,Dh6,Eh6):
    return (sigma_v(x,a,Dv)*Sv(x,a,Dv)+sigma_h6(x,c6,Dh6,Eh6)*Sh6(x,c6,Dh6,Eh6))/sigma6(x,a,c6,Dv,Dh6,Eh6)
def kappae6(x,a,c6,Dv,Dh6,Eh6):
    return ((sigma_h6(x,c6,Dh6,Eh6)*sigma_v(x,a,Dv)*(Sh6(x,c6,Dh6,Eh6)-Sv(x,a,Dv))**2)/(sigma_h6(x,c6,Dh6,Eh6)+sigma_v(x,a,Dv))
            + (kappa_v(x,a,Dv)+kappa_h6(x,c6,Dh6,Eh6)))
def PF6(x,a,c6,Dv,Dh6,Eh6):
    return sigma6(x,a,c6,Dv,Dh6,Eh6)*Seebeck6(x,a,c6,Dv,Dh6,Eh6)**2
def ZT6(x,a,c6,Dv,Dh6,Eh6):
    return (sigma6(x,a,c6,Dv,Dh6,Eh6)*Seebeck6(x,a,c6,Dv,Dh6,Eh6)**2)/kappae6(x,a,c6,Dv,Dh6,Eh6)

npoint = 101
eta = linspace(-10, 10, npoint) 
```

```python
# Create figure object
fig = plt.figure(figsize=(14, 10))

# Add x and y-axes 
axBnd1 = fig.add_axes([0.00, 0.55, 0.30, 0.40])
axDOS1 = fig.add_axes([0.35, 0.55, 0.10, 0.40])


# Add x and y-axis labels 
axBnd1.set_xlabel('')
axBnd1.set_ylabel('Energy (eV)')
axDOS1.set_xlabel('DOS')


# Hide x-axis original tick marks
axBnd1.xaxis.set_tick_params(which='both', bottom=False, top=False)
axDOS1.xaxis.set_tick_params(which='both', bottom=False, top=False)


# Hide x-axis tick labels for DOS
axDOS1.set_xticks([])


# Major tick interval in y-axis every 1 eV
axBnd1.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))
axDOS1.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1))

# Minor tick interval in y-axis every 0.25 eV
axBnd1.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.25))
axDOS1.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.25))


# Edit the major and minor ticks of the x and y axes
axBnd1.yaxis.set_tick_params(which='major', size=6, direction='in', right='on')
axBnd1.yaxis.set_tick_params(which='minor', size=4, direction='in', right='on')
axDOS1.yaxis.set_tick_params(which='major', size=6, direction='in', right='on')
axDOS1.yaxis.set_tick_params(which='minor', size=4, direction='in', right='on')

emin = -2
emax = 2

#################################################################################
prefix = 'so-MoS2' # 
#################################################################################

# Folder location
bndfolder = 'so-MoS2/band/'
dosfolder = 'so-MoS2/dos/'

# Number of bands and k-points from *.bands
nbnd = 62; nks = 301

# Fermi energy, find it in *.scf.out and *.dos
efbnd = -1.6944
efdos = -1.6944

# Open *.bands.gnu file
fbnd = open(bndfolder+prefix+'.bands.gnu', 'r')
f_bnd = [line for line in fbnd.readlines() if line.strip()]
fbnd.close()

# Read k-points
rks = []
for ik in range(nks):
    rks.append(float(f_bnd[ik].split()[0]))

# Set high-symmetry points from *.nscf.in file
rM = rks[0]  ; rG = rks[100] ; rX = rks[200]; rM2 = rks[300]

# Read band index at each k-point
band = []
for ib in range (nbnd):
    band.append([])
    for ik in range(nks):
        tmp = f_bnd[ib*nks+ik].split()[1]
        band[ib].append(float(tmp)-efbnd)
        
# Read eigen-energy at each band index
ener = []
for ik in range(nks):
    ener.append([])
    for ib in range(nbnd):
        tmp = band[ib][ik]
        ener[ik].append(float(tmp))
        
# Open and read DOS data from *.dos file
with open(dosfolder+prefix+'.dos') as fdos:
    next(fdos) # skip header row
    lines = fdos.readlines()
    edos  = np.array([line.split()[0] for line in lines], dtype=float)
    dos   = np.array([line.split()[1] for line in lines], dtype=float)
fdos.close()

#--------------------------------------------------------------------------------
# Set limit and tick labels for high-symmetry points
axBnd1.set_xlim(rM, rM2)
axBnd1.set_xticks([rM, rG, rX, rM2])
axBnd1.set_xticklabels(['M', '$\Gamma$', 'X', 'M'])
axBnd1.set_ylim(emin,emax)
axDOS1.set_xlim(0,0.05)
axDOS1.set_ylim(emin,emax)

# Add dotted lines at middle high-symmetry points
axBnd1.axvline(rM, color='gray', linewidth=1)
axBnd1.axvline(rG, color='gray', linewidth=1)
axBnd1.axvline(rX, color='gray', linewidth=1)
axBnd1.axvline(rM2, color='gray', linewidth=1)

# Add dotted line at Fermi energy
axBnd1.axhline(0, color= 'gray', linestyle='--', linewidth=1)
axDOS1.axhline(0, color= 'gray', linestyle='--', linewidth=1)

# Plot band structure and DOS
axBnd1.plot(rks, ener, color = 'red', linewidth=2)
axDOS1.plot(dos/np.max(dos), edos-efdos, color = 'red', linewidth=2)
axDOS1.fill_betweenx(edos-efdos, 0, dos/np.max(dos), facecolor='salmon', alpha=0.2)


#################################################################################
# Save figure
plt.savefig('FigA10.pdf',bbox_inches='tight')
```

```python
# Import module and function

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

# function for loading BoltzTraP data
def load_data(fname, colnames):
    
    # Find the number of columns. Takes the first line that isn't a comment. 
    with open(fname) as f:
        j=f.readline()
        if j[0] == '#':
            j=f.readline()
        ncols=len(j.split())
    
    # Check that the length of colnames agrees with the output data
    if len(colnames) != ncols:
        print("ERROR:  len(colnames) != ncols.")
        print("The names are not commensurate.")
        print("Is there a different number of columns?")
        print('len(colnames) = ',len(colnames))
        print('ncols = ',ncols)
        return
    
    # Load the file
    d=np.loadtxt(fname, usecols=range(0,ncols) ).T
    
    # Creat a dictionary object, in which a 'key' represents each column name
    data={}
    for col, name in enumerate(colnames):
        data[name] = d[col]
    return data

# Unit conversion
eVtoRy = 0.0734986176
eVtoHa = 0.0734986176 / 2
RytoeV = 1/eVtoRy
HatoeV = 1/eVtoHa

# .trace file column labels
colnames_trace=[ 'mu', 'sigma' , 'S' , 'kappa0' ]

# Create figure object
fig = plt.figure(figsize=(15, 11))

# Add x and y-axes 
axS   = fig.add_axes([0.00, 0.56, 0.44, 0.45])
axSig = fig.add_axes([0.60, 0.56, 0.44, 0.45])
axKap = fig.add_axes([0.00, 0.00, 0.44, 0.45])
axPF  = fig.add_axes([0.60, 0.00, 0.44, 0.45])

# Edit the major and minor ticks of the x and y axes
axS.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axS.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axSig.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axSig.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axKap.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axKap.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')
axPF.xaxis.set_tick_params(which='major', size=8, direction='in', top='on')
axPF.xaxis.set_tick_params(which='minor', size=5, direction='in', top='on')

axS.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axS.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axSig.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axSig.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axKap.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axKap.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')
axPF.yaxis.set_tick_params(which='major', size=8, direction='in', right='on')
axPF.yaxis.set_tick_params(which='minor', size=5, direction='in', right='on')

######################################################
# bulk Cubic
######################################################

## Data location
soMoS2TeC   = 'so-MoS2/dos/RTA'
dataC   = load_data(soMoS2TeC+'.trace', colnames_trace)

#TC300   = np.isclose(dataC['T'], 300)

musoMoS2TeC = dataC['mu']
SsoMoS2TeC  = dataC['S']
SigsoMoS2TeC= dataC['sigma']
KapsoMoS2TeC= dataC['kappa0']

EfsoMoS2TeC = -1.6944

#----------------------------------------------------------------------
# Plotting
#----------------------------------------------------------------------

#----------------------------------------------------------------------
# Seebeck coefficient
#----------------------------------------------------------------------

axS.plot(musoMoS2TeC-EfsoMoS2TeC, SsoMoS2TeC*1e6, color='red', linestyle='-',
         linewidth=2.0, label='Cubic, $T =$ 300 K')


#----------------------------------------------------------------------
# Electrical conductivity
#----------------------------------------------------------------------

axSig.plot(musoMoS2TeC-EfsoMoS2TeC, 16.22*SigsoMoS2TeC*1e-6, color='red', linestyle='-',
         linewidth=2.0, label='cubic, T = 300 K')

#----------------------------------------------------------------------
# Electronic thermal conductivity
#----------------------------------------------------------------------

axKap.plot(musoMoS2TeC-EfsoMoS2TeC, 16.22*KapsoMoS2TeC*1e0, color='red', linestyle='-',
         linewidth=2.0, label='cubic, T = 300 K')

#----------------------------------------------------------------------
# PF
#----------------------------------------------------------------------

axPF.plot(musoMoS2TeC-EfsoMoS2TeC, 100*16.22*SsoMoS2TeC**2*SigsoMoS2TeC, color='red', 
          linestyle='-', linewidth=2.0, label='cubic, T = 300 K')


#######################################################################
# Format Axis Labels and Ticks
#######################################################################

axS.axhline(0, color= 'grey', linestyle='-', linewidth=1)
axS.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axS.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.2))
axS.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.1))
axS.yaxis.set_major_locator(mpl.ticker.MultipleLocator(100))
axS.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(50))
axS.set_xlim(-0.4,0.4); axS.set_ylim(-200,200)
axS.tick_params(axis = 'x', labelsize = 38)
axS.tick_params(axis = 'y', labelsize = 38)
axS.set_ylabel(r'$S$ ($\mu$V/K)', labelpad=0, fontsize = 45)

axSig.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axSig.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.2))
axSig.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.1))
axSig.yaxis.set_major_locator(mpl.ticker.MultipleLocator(3))
axSig.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(1.5))
axSig.set_xlim(-0.4,0.4); axSig.set_ylim(0,9)
axSig.tick_params(axis = 'x', labelsize = 38)
axSig.tick_params(axis = 'y', labelsize = 38)
axSig.set_ylabel(r'$\sigma$ ($10^6$/$\Omega$.m.s)', labelpad=12, fontsize = 45)

axKap.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axKap.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.2))
axKap.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.1))
axKap.yaxis.set_major_locator(mpl.ticker.MultipleLocator(20))
axKap.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(10))
axKap.set_xlim(-0.4,0.4); axKap.set_ylim(0,65)
axKap.tick_params(axis = 'x', labelsize = 38)
axKap.tick_params(axis = 'y', labelsize = 38)
axKap.set_ylabel(r'$\kappa_e$ (W/m.K.s)', labelpad=16, fontsize = 45)
axKap.set_xlabel('Fermi Energy (eV)', labelpad = 16, fontsize = 45)

axPF.axvline(0, color= 'grey', linestyle='-', linewidth=1)
axPF.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.2))
axPF.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.1))
axPF.yaxis.set_major_locator(mpl.ticker.MultipleLocator(1.0))
axPF.yaxis.set_minor_locator(mpl.ticker.MultipleLocator(0.5))
axPF.set_xlim(-0.4,0.4); axPF.set_ylim(0,4)
axPF.tick_params(axis = 'x', labelsize = 38)
axPF.tick_params(axis = 'y', labelsize = 38)
axPF.set_ylabel(r'PF ($10^{-2}$W/m.$\mathregular{K^2}$)', labelpad=10, fontsize = 45)
axPF.set_xlabel('Fermi Energy (eV)', labelpad = 16,fontsize = 45)

## Give panel labels
axS.text(-0.65,205, '(a)', fontsize = 42)
axSig.text(-0.65, 9.1, '(b)', fontsize =42 )
axKap.text(-0.65, 66, '(c)', fontsize = 42)
axPF.text(-0.65, 4.1, '(d)', fontsize = 42)

#################################################################################
# Save figure
plt.savefig('FigA11.pdf',bbox_inches='tight')
```
